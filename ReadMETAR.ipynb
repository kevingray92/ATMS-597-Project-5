```
import numpy as np
import pandas as pd
from datetime import datetime
import metpy
import metpy.io
```

```
###Download the METAR data
!wget ftp://ftp.ncdc.noaa.gov/pub/data/asos-fivemin/6401-2000/64010KMSN200001.dat
```

```
###Read the METAR file. An error will result after the last line, but the data should be saved to the decoded_metar DataFrame.
with open('64010KMSN200001.dat','r') as metar_file:
    metar_line = metar_file.readline()
    metar_line_clean = metar_line[53:] #Get rid of first 53 characters in each line, superfluous info
    decoded_metar = metpy.io.parse_metar_to_dataframe(metar_line_clean, year=2000, month=1) #decode the METAR
    cnt = 1 #Start a counter
    while metar_line:
        metar_line = metar_file.readline()
        metar_line_clean = metar_line[53:]
        decoded_metar_line = metpy.io.parse_metar_to_dataframe(metar_line_clean, year=2000, month=1)
        decoded_metar_line.set_index(keys='date_time')
        decoded_metar = decoded_metar.append(decoded_metar_line)
        decoded_metar
        cnt = cnt+1
```

```
###Make date_time the index
#decoded_metar = decoded_metar.set_index(keys='date_time')
decoded_metar = decoded_metar.set_index(pd.to_datetime(decoded_metar.index))
decoded_metar
```
